```python
def generatetab(n):
    table =""
    for i in range (1,11):
        table+= f"{n} X {i} + {n*i} \n"

    with open (f"Tables/table_{n}","w") as f :
        f.write (table)
        


for i in range (2,21):
    generatetab(i)
```

```python
def gentab(n):
    table =""
    for i in range (1,11):
        table+=f"{n} X {i} = {n*i} \n"
    with open(f"Tables/table_{n}","w") as f:
        f.write (table )
for i in range (2,31):
    gentab(i)
```
