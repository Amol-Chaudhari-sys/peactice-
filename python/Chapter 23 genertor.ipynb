```python
from functools import lru_cache
import time 
```

```python
@lru_cache(maxsize=None)
def fx(n):
    time.sleep(5)
    return n*5
```

```python
print(fx(20))
print(fx(4))
print(fx(8))
print(fx(20))
print(fx(4))
print(fx(8))


print(fx(21))
print(fx(4))
print(fx(8))
print(fx(22))
print(fx(4))
print(fx(7))
```
