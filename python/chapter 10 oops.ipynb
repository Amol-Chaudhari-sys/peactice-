```python
class Employe:
    lang="java"
    age =12
    sal = 999
    def __init__(self,name ,lan ,ag,sl ):  # dunder method which i sautomaticaly called starts With ( __ )
        print ("i am constructor ")
        self.name =name 
        self.lang=lan 
        self.age =ag
        self.sal = sl
    
    def getinfo(self):
        print (f"name : {self.name }\n language : {self.lang } \n age :{self.age }\n salary : {self.sal}")
        
    @staticmethod #  if no vaue taken from any class or object
    def greet( ):
        print (f"hello ")
```

```python
def greet(fn):
    def mfx():
        print ("welcome in greet ")
        fn()
        print ("function run ")
    return mfx

@greet
def h():
    print ("hello world")
h()
```

```python
class math:
    def __init__(self ,num):
        self.num =num
    def addton(self,n):
        self.num=self.num+n

    @staticmethod
    def addtwo(a,b):
        return a+b




a=math(10)
print (a.num)
a.addton(12)
print (a.num)
a.addtwo(44,13)
```
