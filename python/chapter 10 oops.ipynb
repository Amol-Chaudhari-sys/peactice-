```python
class Employe:
    lang="java"
    age =12
    sal = 999
    def __init__(self,name ,lan ,ag,sl ):  # dunder method which i sautomaticaly called starts With ( __ )
        print ("i am constructor ")
        self.name =name 
        self.lang=lan 
        self.age =ag
        self.sal = sl
    
    def getinfo(self):
        print (f"name : {self.name }\n language : {self.lang } \n age :{self.age }\n salary : {self.sal}")
        
    @staticmethod #  if no vaue taken from any class or object
    def greet( ):
        print (f"hello ")
```

```python
hari = Employe()
hari.name ="harri"
print(hari.sal, hari.name)   
amol=Employe()
amol.name = "Amol"
amol.lang ="python"
hari.getinfo()    ####
print (amol.name)
```

```python
amol.greet()
```
