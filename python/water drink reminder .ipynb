```python
import os

title = "Reminder"
message = "Take a short break!"

# PowerShell command for toast notification
os.system(f'powershell -Command "New-BurntToastNotification -Text \'{title}\', \'{message}\'"')
```

```python
import os 
import time 
from plyer import notification
import pyttsx3
engine = pyttsx3.init()
engine.setProperty("rate", 150)
```

```python
i=0
text="hey drink some water "
while (i<6) :
    notification.notify(
            title="Water" ,
            message="Hey drink some water ",
            timeout=3  # Duration in seconds
        )
    engine.say(text)
    engine.runAndWait()    
    time .sleep(5)
    print (i)
    i+=1
```
