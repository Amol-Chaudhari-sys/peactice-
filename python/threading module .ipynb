```python
time1= time.perf_counter()

t1= threading.Thread(target= fun , args = [1])
t2= threading.Thread(target= fun , args = [2])
t3= threading.Thread(target= fun , args = [3])
t1.start()
t2.start()
t3.start()
t1.join()
t2.join ()
t3.join ()


time2 = time.perf_counter()
print (time2-time1)
```

```python
import multiprocessing 
import requests
import concurrent.futures
```

```python
def download (url , name):
    print (f"start downloading {name}")
    response =requests.get(url)
    open(f"files2/file{name}.jpg", "wb").write(response.content)
    print (f"finish downloading {name }")
    
```

```python
if __name__ == "__main__":
    url = "https://picsum.photos/200/300"
    name = "dog"
    pros=[]
    
    for i in range (5):
        p = multiprocessing.Process(target=download, args =[url, i ]) 
        p.start()
        pros.append (p)
    
    for p in pros:
        p.join ()
```

```python
url = "https://picsum.photos/200/300"
with concurrent.futures.ProcessPoolExecutor() as executer :
    l1= [i for i in range (5)]
    l2= [url for i in range (5)]
    result = executer.map(download ,l2,l1)
    for i in result :
        print (i)
```
