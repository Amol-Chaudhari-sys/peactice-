```python
def fact(n):
    if (n==1 or n==0):
        return 1
   
    return n*fact(n-1)

```

```python
fact(340)
```

```python
def great(a,b,c):
    if (a>b and a>c ):
        return a 
    elif (b>a and b>c):
        return b

    else : return c
```

```python
a=great(25,76,40)
print (f"{a} degree  celcious")
```

```python
def sum(n):
    if (n==1):
        return 1
    return n+sum(n-1)
```

```python
sum(5)
```

```python
def patt(n):
    if (n==0):
        return 
    print(" ^ " *(n),end ="")
    print("")
    patt(n-1)
```

```python
l1= ['  amol  ', '   arnav','soham','sujal']
def rmo(n,l):
    m=[]
    for i in l:
        if not (i==n):
            m.append(i.strip(" "))
    return m
    
```

```python
a=rmo('arnav',l1)
print (a)
print(l1)
```
