```python
def fact(n):
    if (n==1 or n==0):
        return 1
   
    return n*fact(n-1)

```

```python
fact(5)
```

```python
def great(a,b,c):
    if (a>b and a>c ):
        return a 
    elif (b>a and b>c):
        return b

    else : return c
```

```python
a=great(25,76,40)
print (f"{a} degree  celcious")
```

```python
l1= ['  amol  ', '   arnav','soham','sujal']
def rmo(n,l):
    m=[]
    for i in l:
        if not (i==n):
            m.append(i.strip(" "))
    return m
    
```

# enumerate function in python 

```python
# ordinary method 
marks =[54,87,56,87,9,36,27,87,45,79]
index =0
for i in marks:
    print (i)
    if index ==4:
        print ("this is 4 th index ")
    index +=1
        
```

```python
# enumerate method 
marks =[54,87,56,87,9,36,27,87,45,79]

for index,i in enumerate( marks,start =1):
    print (index,i)
    if index ==4:
        print ("this is 4 th index ")
    
```

# local and global veriable

```python
x= 10
def fun():
    global x
    x=8
    print(x)
fun()
print (x)
```

# annonumous function lambda function 

```python
double =lambda x :x*2
cube =lambda x:x*x*x
avg =lambda x,y:(x+y)/2
```

```python
print(double (10))
print(cube (9))
print (avg (10,20))
```

```python
def app(fx,f):
    return 4+fx(f)

app(lambda x:x*x,4)  # annomous function 
```
