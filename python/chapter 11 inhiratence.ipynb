```python
class comp:
    def __init__(self,r,i):
        self.r=r
        self.i =i
    def __add__(self,c2):
        return (self.r+c2.r ,self.i+c2.i)

    def __mul__(self,c2):
        real=self.i*c2.i-self.r*c2.r
        img=self.r*c2.i+self.i*c2.r
        return comp(real,img)

    def __str__(self):
        return f"{self.r} + {self.i}i"

a=comp(2,3)
b=comp(3,4)
c=(a+b)
print (a*b)
print (c)
```

```python
class vector :
    def __init__ (self,x,y,z ):
        self.x=x
        self.y=y
        self.z=z
    def __add__(self,v2 ):
        result=self.x+v2.x,self.y+v2.y,self.z+v2.z
        return (result)

    def __mul__(self,v2):
        result=self.x*v2.x+self.y*v2.y+self.z*v2.z
        return result
    def __str__(self):
        return f"vector({self.x}i+{self.y}j+{self.z}k) "
        
```

```python
v1=vector(2,3,4)
v2=vector(5,6,7)

print (v1+v2)
print (v1*v2) 
```

```python
class shape :
    def __init__(self ,b,l):
        self.breadth =b
        self.length =l
    def area(self ):
        return self.length *self.breadth 
class circle(shape ) :
    def __init__(self ,r ):
        self.radius =r
    def area(self ):
        return 3.14*self.radius*3.14
```

```python
a=circle(10)
a.area()
```
