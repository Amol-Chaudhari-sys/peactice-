```python
# import pywin32

import pyttsx3

```

```python
# speaker = win32com.client.Dispatch("SAPI.SpVoice")
engine = pyttsx3.init()
engine.setProperty("rate", 150)
```

```python
l=['amol','soham','walture','vikas','arnav','aishwarya','vedika']
```

```python
for name in l :
    text=f'my name is {name}'
    # speaker.Speak(text)
    engine.say(text)

engine.runAndWait()
```
