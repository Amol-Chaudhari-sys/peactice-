```python
import pandas as pd 
import numpy as np 
import seaborn as sns
# import matplotlab.pyplot as plt
```

```python
pop_data=[np.random.randint(10,100) for i in range (10000)]
pop_table=pd.DataFrame({"pop":pop_data})
pop_table
```

```python
sns.kdeplot(x="pop",data=pop_table)
```

```python
sample_mean=[]
for no_sample in range (10):
    sample=[]
    for data in range (500):
        sample.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample))
    
```

```python
samp_M=pd.DataFrame({"mean":sample_mean})
```

```python
sns.kdeplot(x="mean",data=samp_M)
```
