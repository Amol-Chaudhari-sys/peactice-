# z-Test

```python
import scipy.stats as st
import numpy as np 

p_u=82
p_std=20
s_x=90
no_s=81
ap=0.05

z_cal= (s_x-p_u)/(p_std/np.sqrt(no_s))
print (z_cal)
z_table=st.norm.ppf(1-ap)
print(z_table)
```

```python
if z_table <z_cal:
    print ("ha is right "+ str(z_cal))
else :
    print ("H0 is right ")
```
